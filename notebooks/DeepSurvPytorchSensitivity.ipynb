```python
import matplotlib.pyplot as plt
import numpy as np
import sys
import os
import pickle
sys.path.append('/home/jevjev/Dropbox/Projects/DeepSurv')
```

```python
results = '/home/jevjev/Dropbox/Projects/DeepSurv/sensitivity_results'
```

```python
results_dictionary = {}
for root, dirr, files in os.walk(results):
    if files:
        dataset_name = root.split("/")[-2]
        with open(os.path.join(root, '{}results.pickle'.format(600)), 'rb') as f:
            out = pickle.load(f)
        results_dictionary[dataset_name] = out
print(results_dictionary.keys())
```

```python
fig, ax = plt.subplots()
for dataset_name, output in results_dictionary.items():
    ax.plot(output[0], [i * -1 for i in output[1]], label=dataset_name)
plt.xlim([0,100])
ax.set_xticks(output[0])
plt.xlabel('Batch Size')
plt.legend()
plt.show()
```
